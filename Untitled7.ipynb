```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

student = np.array([[168,10,1],
     [178,5,1],
     [157,30,0],
     [161,40,0],
     [190,70,1],
     [143,5,0]])

male_student = student[student[:,2]==1]
female_student = student[student[:,2]==0]

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure(1)
plt.subplot(221)
plt.plot(t1,f(t1),'bo', t2,f(t2),'k')

plt.subplot(222)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')

plt.subplot(223)
plt.scatter(male_student[:,0],male_student[:,1],marker="^")
plt.scatter(female_student[:,0],female_student[:,1])

plt.show()
```

```python
f,ax=plt.subplots(3,4)
ax[0,0].plot([1,2,3,434,2,343,3,23,4])
ax[1,2].scatter(male_student[:,0],male_student[:,1],marker="^")
ax[2,3].scatter(female_student[:,0],female_student[:,1],marker="^")
plt.show()
```
